```python
from scipy.stats import ks_2samp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
from os import listdir
from os.path import isfile, join
import decimal
from kneed import KneeLocator, DataGenerator

%matplotlib inline

sns.set(style='white', context='notebook', rc={'figure.figsize':(14,10)})
```

```python
mypath = './data'
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
```

```python
datafiles = [f for f in listdir(mypath) if isfile(join(mypath, f)) if f.endswith('.csv')]
```

```python
datafiles
```

```python
dfs = []

for f in datafiles:
    
    df = pd.read_csv(mypath + '/' + f)
    df = df.dropna()
    df = df.select_dtypes(include=numerics)
    
    dfs.append(df)
```

```python
df_test = dfs[1]
```

```python
len(df_test)
```

```python
fractions = [.9, .8, .7, .6, .5, .4, .3, .2, .1]
if len(df_test) < 5000:
    alpha = 0.05/(np.sqrt(((len(df_test) + len(df_test))/(len(df_test) * len(df_test)))))
else:
    alpha = 1.037
stats_fractions = []
pvals_fractions = []
ks_vals = []

for f in fractions:
    df_frac = df_test.sample(frac=f)    

    stats = []
    pvals = []
    for c in df_test.columns:
        kst = ks_2samp(df_test[c].values, df_frac[c].values, mode='asymp')
        stats.append(kst[0])
        pvals.append(1 - kst[1])
        
    stats_fractions.append(stats)
    pvals_fractions.append(pvals)
    ks_val = alpha*(np.sqrt(((len(df_frac) + len(df_test))/(len(df_frac) * len(df_test)))))
    ks_vals.append(ks_val)
```

```python
stats_fractions = np.asarray(stats_fractions)
pvals_fractions = np.asarray(pvals_fractions)
```

```python
alpha
```

```python
for i,v in enumerate(df_test.columns):
    plt.plot(stats_fractions[:, i])
    plt.xticks(range(10), fractions)
    #plt.hlines(0.05, colors='r', linestyles='dashed', xmin=0.0, xmax=8.0)
    plt.title(datafiles[0])
    if len(df_test.columns) < 10:
        plt.legend(df_test.columns)
plt.plot(ks_vals, color='r', linestyle = 'dotted')
```

```python
for i,v in enumerate(df_test.columns):
    plt.plot(pvals_fractions[:, i])
    plt.xticks(range(10), fractions)
    #plt.yscale('log')
    plt.title(datafiles[0])
    if len(df_test.columns) < 10:
        plt.legend(df_test.columns)
plt.hlines(0.05, colors='r', linestyles='dotted', xmin=0.0, xmax=8.0)
```

```python
stats_mean = np.mean(stats_fractions, axis=1)
pvals_mean = np.mean(pvals_fractions, axis=1)
```

```python
stats_mean
```

```python
plt.plot(stats_mean)
plt.plot(ks_vals, color='r', linestyle = 'dotted')
plt.xticks(range(10), fractions)
plt.title(datafiles[0] + ' KS stats')
kneedle = KneeLocator(stats_mean, range(9), S=1.0, curve='convex', direction='increasing')
kneedle.plot_knee_normalized()
```

```python
fractions= np.linspace(1.0, 0.1, num=10)

for idx, df in enumerate(dfs):
    
    if len(df) < 5000:
        alpha = 0.05/(np.sqrt(((len(df) + len(df))/(len(df) * len(df)))))
    else:
        alpha = 1.037
    stats_fractions = []
    pvals_fractions = []
    ks_vals = []

    for f in fractions:
        df_frac = df.sample(frac=f)    

        stats = []
        pvals = []
        for c in df.columns:
            kst = ks_2samp(df[c].values, df_frac[c].values, mode='asymp')
            stats.append(kst[0])
            pvals.append(1 - kst[1])

        stats_fractions.append(stats)
        pvals_fractions.append(pvals)
        ks_val = alpha*(np.sqrt(((len(df_frac) + len(df))/(len(df_frac) * len(df)))))
        ks_vals.append(ks_val)
        

    stats_fractions = np.asarray(stats_fractions)
    pvals_fractions = np.asarray(pvals_fractions)
    
    for i,v in enumerate(df.columns):
        plt.plot(stats_fractions[:, i])
        plt.xticks(range(10), [np.round(f, 1) for f in fractions])
        #plt.hlines(0.05, colors='r', linestyles='dashed', xmin=0.0, xmax=8.0)
        plt.title(datafiles[0])
        if len(df.columns) < 10:
            plt.legend(df.columns)
    plt.plot(ks_vals, color='r', linestyle = 'dotted')
    plt.savefig('./outputs/'+datafiles[idx]+' KS stats all'+'.pdf', bbox_inches='tight')
    plt.show()
    
    for i,v in enumerate(df.columns):
        plt.plot(pvals_fractions[:, i])
        plt.xticks(range(10), [np.round(f, 1) for f in fractions])
        #plt.yscale('log')
        plt.title(datafiles[0])
        if len(df.columns) < 10:
            plt.legend(df.columns)
    plt.hlines(0.05, colors='r', linestyles='dotted', xmin=0.0, xmax=8.0)
    plt.savefig('./outputs/'+datafiles[idx]+' pvals all'+'.pdf', bbox_inches='tight')
    plt.show()
    
    stats_mean = np.mean(stats_fractions, axis=1)
    pvals_mean = np.mean(pvals_fractions, axis=1)
    
    plt.plot(stats_mean)
    plt.plot(ks_vals, color='r', linestyle = 'dotted')
    plt.xticks(range(10), [np.round(f, 1) for f in fractions])
    plt.title(datafiles[0] + ' KS stats mean')
    plt.savefig('./outputs/'+ datafiles[idx]+ ' KS stats mean'+'.pdf', bbox_inches='tight')
    plt.show()
    
    kneedle = KneeLocator(stats_mean, range(10), S=1.0, curve='convex', direction='increasing')
    kneedle.plot_knee_normalized()
    plt.show()
    
    plt.plot(pvals_mean)
    plt.xticks(range(10), [np.round(f, 1) for f in fractions])
    plt.title(datafiles[0] + ' p-values mean')
    plt.hlines(0.05, colors='r', linestyles='dotted', xmin=0.0, xmax=8.0)
    plt.savefig('./outputs/'+datafiles[idx]+' KS stats mean'+'.pdf', bbox_inches='tight')
    plt.show()
    
    kneedle = KneeLocator(pvals_mean, range(10), S=1.0, curve='convex', direction='increasing')
    kneedle.plot_knee_normalized()
    plt.show()
```

```python
def singleDistributionTest(path_in='./data', path_out='./outputs'):
    
    fractions= np.linspace(1.0, 0.1, num=10)
    
    mypath = path_in
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    
    datafiles = [f for f in listdir(mypath) if isfile(join(mypath, f)) if f.endswith('.csv')]
    
    print(datafiles)
    
    dfs = []

    for f in datafiles:

        df = pd.read_csv(mypath + '/' + f)
        df = df.dropna()
        df = df.select_dtypes(include=numerics)

        dfs.append(df)

    for idx, df in enumerate(dfs):

        if len(df) < 5000:
            alpha = 0.05/(np.sqrt(((len(df) + len(df))/(len(df) * len(df)))))
        else:
            alpha = 1.037
        stats_fractions = []
        pvals_fractions = []
        ks_vals = []

        for f in fractions:
            df_frac = df.sample(frac=f)    

            stats = []
            pvals = []
            for c in df.columns:
                kst = ks_2samp(df[c].values, df_frac[c].values, mode='asymp')
                stats.append(kst[0])
                pvals.append(1 - kst[1])

            stats_fractions.append(stats)
            pvals_fractions.append(pvals)
            ks_val = alpha*(np.sqrt(((len(df_frac) + len(df))/(len(df_frac) * len(df)))))
            ks_vals.append(ks_val)


        stats_fractions = np.asarray(stats_fractions)
        pvals_fractions = np.asarray(pvals_fractions)

        for i,v in enumerate(df.columns):
            plt.plot(stats_fractions[:, i])
            plt.xticks(range(10), [np.round(f, 1) for f in fractions])
            #plt.hlines(0.05, colors='r', linestyles='dashed', xmin=0.0, xmax=8.0)
            plt.title(datafiles[0])
            if len(df.columns) < 10:
                plt.legend(df.columns)
        plt.plot(ks_vals, color='r', linestyle = 'dotted')
        plt.savefig(path_out+'/'+datafiles[idx]+' KS stats all'+'.pdf', bbox_inches='tight')
        plt.show()

        for i,v in enumerate(df.columns):
            plt.plot(pvals_fractions[:, i])
            plt.xticks(range(10), [np.round(f, 1) for f in fractions])
            #plt.yscale('log')
            plt.title(datafiles[0])
            if len(df.columns) < 10:
                plt.legend(df.columns)
        plt.hlines(0.05, colors='r', linestyles='dotted', xmin=0.0, xmax=9.0)
        plt.savefig(path_out+'/'+datafiles[idx]+' pvals all'+'.pdf', bbox_inches='tight')
        plt.show()

        stats_mean = np.mean(stats_fractions, axis=1)
        pvals_mean = np.mean(pvals_fractions, axis=1)

        plt.plot(stats_mean)
        plt.plot(ks_vals, color='r', linestyle = 'dotted')
        plt.xticks(range(10), [np.round(f, 1) for f in fractions])
        plt.title(datafiles[0] + ' KS stats mean')
        plt.savefig(path_out+'/'+ datafiles[idx]+ ' KS stats mean'+'.pdf', bbox_inches='tight')
        plt.show()
        
        kneedle = KneeLocator(stats_mean, range(10), S=1.0, curve='convex', direction='increasing')
        kneedle.plot_knee_normalized()
        plt.show()

        plt.plot(pvals_mean)
        plt.xticks(range(10), [np.round(f, 1) for f in fractions])
        plt.title(datafiles[0] + ' p-values mean')
        plt.hlines(0.05, colors='r', linestyles='dotted', xmin=0.0, xmax=9.0)
        plt.savefig(path_out+'/'+datafiles[idx]+' KS stats mean'+'.pdf', bbox_inches='tight')
        plt.show()
        
        kneedle = KneeLocator(pvals_mean, range(10), S=1.0, curve='convex', direction='increasing')
        kneedle.plot_knee_normalized()
        plt.show()
```

```python
def doubleDistributionTest(df1, df2, path_in='./data', path_out='./outputs'):
    
    fractions= np.linspace(1.0, 0.1, num=10)
    
    mypath = path_in
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    
    #datafiles = [f for f in listdir(mypath) if isfile(join(mypath, f)) if f.endswith('.csv')]
    
    dfs = []

    for f in [df1, df2]:
        
        df = pd.read_csv(mypath + '/' + f)
        df = df.dropna()
        df = df.select_dtypes(include=numerics)

        dfs.append(df)

    if len(dfs[0]) < 5000:
        alpha = 0.05/(np.sqrt(((len(dfs[0]) + len(dfs[0]))/(len(dfs[0]) * len(dfs[0])))))
    else:
        alpha = 1.037
        
    stats_fractions = []
    pvals_fractions = []
    ks_vals = []

    for f in fractions:
        df_frac = dfs[1].sample(frac=f)    

        stats = []
        pvals = []
        for c in dfs[1].columns:
            kst = ks_2samp(dfs[0][c].values, df_frac[c].values, mode='asymp')
            stats.append(kst[0])
            pvals.append(1 - kst[1])

        stats_fractions.append(stats)
        pvals_fractions.append(pvals)
        ks_val = alpha*(np.sqrt(((len(df_frac) + len(dfs[0]))/(len(df_frac) * len(dfs[0])))))
        ks_vals.append(ks_val)


    stats_fractions = np.asarray(stats_fractions)
    pvals_fractions = np.asarray(pvals_fractions)

    for i,v in enumerate(dfs[0].columns):
        plt.plot(stats_fractions[:, i])
        plt.xticks(range(10), [np.round(f, 1) for f in fractions])
        #plt.hlines(0.05, colors='r', linestyles='dashed', xmin=0.0, xmax=8.0)
        plt.title(df1+' KS stats all')
        if len(dfs[0].columns) < 10:
            plt.legend(dfs[0].columns)
    plt.plot(ks_vals, color='r', linestyle = 'dotted')
    plt.savefig(path_out+'/'+df1+' KS stats all'+'.pdf', bbox_inches='tight')
    plt.show()

    for i,v in enumerate(dfs[0].columns):
        plt.plot(pvals_fractions[:, i])
        plt.xticks(range(10), [np.round(f, 1) for f in fractions])
        #plt.yscale('log')
        plt.title(df1+' pvals all')
        if len(dfs[0].columns) < 10:
            plt.legend(dfs[0].columns)
    plt.hlines(0.05, colors='r', linestyles='dotted', xmin=0.0, xmax=9.0)
    plt.savefig(path_out+'/'+df1+' pvals all'+'.pdf', bbox_inches='tight')
    plt.show()

    stats_mean = np.mean(stats_fractions, axis=1)
    pvals_mean = np.mean(pvals_fractions, axis=1)

    plt.plot(stats_mean)
    plt.plot(ks_vals, color='r', linestyle = 'dotted')
    plt.xticks(range(10), [np.round(f, 1) for f in fractions])
    plt.title(df1 + ' KS stats mean')
    plt.savefig(path_out+'/'+ df1+ ' KS stats mean'+'.pdf', bbox_inches='tight')
    plt.show()

    plt.plot(pvals_mean)
    plt.xticks(range(10), [np.round(f, 1) for f in fractions])
    plt.title(df[0] + ' p-values mean')
    plt.hlines(0.05, colors='r', linestyles='dotted', xmin=0.0, xmax=9.0)
    plt.savefig(path_out+'/'+df1+' KS stats mean'+'.pdf', bbox_inches='tight')
    plt.show()
```
